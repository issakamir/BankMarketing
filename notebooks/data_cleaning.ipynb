## 1. Data Cleaning & Pre-processing
In this section, we prepare the raw bank marketing data for analysis. This includes handling missing values, duplicates, and converting categorical strings into numeric formats for statistical modeling.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### 1.2 Binary Encoding
Several columns contain 'yes' or 'no' strings (e.g., deposit, housing, loan). We convert these into binary integers (1 and 0) so we can calculate means, correlations, and success rates.

```python
cols_to_fix=['default', 'deposit', 'loan', 'housing']
for col in cols_to_fix:
    df[f'{col}_num']=df[col].map({'yes':1, 'no':0})

df=df.drop(columns=cols_to_fix)
df.head()
```

### 1.3 Handling Temporal Data
The 'month' column is currently a string. We convert it to an ordered Categorical type to ensure that visualizations follow the chronological order of the calendar year rather than alphabetical order.

```python
month_order=['jan', 'feb', 'mar','apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df['month']=pd.Categorical(df['month'], categories=month_order, ordered=True)
df.head()
```

### 1.4 Duration to minutes
The raw 'duration' of the calls was recorded in seconds. We transformed this into a new feature, duration_min, to make the data more interpretable for business reporting. The original 'duration' column was then dropped to maintain a lean dataset.

```python
df['duration_min']=df['duration']/60
df=df.drop(columns=['duration'])
df.head()
```
